 ## **a)** Si tomamos `BB` como variable dependiente seleccione 2 variables independientes utilizando el método ‘hacia adelante’ (forward) para hacer una regresión lineal múltiple. Apóyese de un software matemático para hacer la regresión.
 

>En este caso tomaremos a las variables `BF` y `IP` como las variables independientes y a `BB` como la variable dependiente

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("2022 MLB Player Stats - Pitching.csv",sep = ';')

X = data[["IP","BF"]] # variables independientes
y = data["BB"] # variable dependiente

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

 ## **b)** De la regresión anterior compruebe los supuestos. Cuál no se cumple? Cómo se interpreta esto geométricamente?
 


### **Supuesto #1:** Existe correlación lineal entre las variables dependientes e independientes

```python
data = pd.read_csv("2022 MLB Player Stats - Pitching.csv",sep = ';')
data = data[["IP","BF","BB","Tm"]]
corr_matrix =  data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
sns.pairplot(data = data,hue = "Tm")
plt.show()
```

### **Supuesto #2:** Los errores son independientes

```python
plt.scatter(model.predict(), model.resid)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
# -----errores independientes-----------
# Calculamos el estadistico de Durbin-Watson 
dw = sm.stats.stattools.durbin_watson(model.resid)

print('Durbin-Watson statistic:', dw)
```

>Si el valor estadístico de Durbin-Watson está cerca de 2, entonces no hay una autocorrelación significativa en los residuos y los errores pueden considerarse independientes.

>En este caso el test nos dio `1.95093073315656`. 
>Por lo cual podemos decir que son independientes los errores

### **Supuesto #3:** El valor esperado del error aleatorio es 0 y la suma de estos también es 0

```python
#-------valor esperado del error aleatorio e_i es 0----------------
mean = sum(model.resid)/len(model.resid)
suma = sum(model.resid)
alpha = 0.05
if mean < alpha:
    print(f"La esperanza de los errores es cero (mean={mean}, alpha={alpha})")
else:
    print(f"La esperanza de los errores no es cero (mean={mean}, alpha={alpha})")
    
if suma < alpha:
    print(f"La suma de los errores es cero (suma={suma}, alpha={alpha})")
else:
    print(f"La suma de los errores no es cero (suma={suma}, alpha={alpha})")
```

```python
#----------Homocedasticidad----------------------
import statsmodels.stats.diagnostic as smd

# Realizar la prueba de Breusch-Pagan
bp_test = smd.het_breuschpagan(model.resid, X)
p_value_bp = bp_test[1]

# Realizar la prueba de White
white_test = smd.het_white(model.resid, X)
p_value_w = white_test[1]

#También se podria utilizar el p-value del test de white para este supuesto

alpha = 0.05
if bp_test[1] < alpha:
    print("Cumple con la Homocedasticidad")
else:
    print("Hay Heterocedasticidad")
```

## **a)** Las medias de la proporción `SO/W` de los equipos `OAK`, `SEA`, `TEX`, y `NYY` se pueden considerar estadísticamente iguales? (Recuerda comprobar los supuestos)

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as sc
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson
import seaborn as sns


data = pd.read_csv("2022 MLB Player Stats - Pitching.csv", sep=';')
data = data[data['Lg'] == 'AL']
data = data[data['Tm'].isin(['OAK', 'SEA', 'TEX','NYY'])]
data = data.rename(columns={'SO/W': 'SO_W'})

oak = data['SO_W'][data['Tm'] == 'OAK']
sea = data['SO_W'][data['Tm'] == 'SEA']
tex = data['SO_W'][data['Tm'] == 'TEX']
nyy = data['SO_W'][data['Tm'] == 'NYY']

var = [oak,sea,tex,nyy]

 # Crear modelo ANOVA
model = ols('SO_W ~ Tm', data=data).fit()
 # Obtener tabla ANOVA
table = sm.stats.anova_lm(model, typ=2)
 # Imprimir tabla ANOVA
print(table)

residuos = model.resid
 
```

## **b)** Haga una reducción de dimensión sobre las características de los jugadores que participaron en la Liga Americana, usando K-Means. Agrupe en 4 clústeres.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("2022 MLB Player Stats - Pitching.csv", sep=';')
data = data[data['Lg'] == 'AL']
data = data.drop(['Rk', 'Name','Age','Tm','Lg'], axis=1)


# Estandarizamos los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

#Hacemos la reduccion de dimensiones
pca = PCA()
pca.fit(data_scaled)

# Calculate the cumulative explained variance ratio
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)

# Select the number of components that explain at least % of the variance    
n_components = np.argmax(cumulative_var_ratio >= 0.85) + 1

print("\n\nNumber of components:", n_components)
 # Transform the data using the selected number of components
data_pca = pca.transform(data_scaled)[:, :n_components]


# View the selected components
components = pca.components_
print("\n\nSelected components:")
#print(components)


#Aplicamos el algoritmo K-Means para agrupar los datos en 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit_predict(data_pca)

# Graficamos los clusters
plt.scatter(data_pca[:,0], data_pca[:,1], c=clusters, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

# Algunos de los test usados:

### **Test de White (Prueba de Heterocedasticidad)**

**Ho:** Existe heterocedasticidad en los residuos  

**Ha:** Existe homocedasticidad en los residuos `(varianza constante)`

### **Test de Jerque-Bera (Prueba de Normalidad)**

**Ho:** Los residuos distribuyen normal  

**Ha:** Los residuos no distribuyen normal

### **Test de Durbin-Watson (Prueba de Autocorrelación)**

**Ho:** $\rho = 0$

**Ha:** $\rho > 0$

$~Si~~ d ~<~ d_L~~ Rechazamos~ Ho$  
$Si ~~d~ >~ d_U~~ No~ Rechazamos~ Ho$   
$Si ~~d_L ~< ~d ~<~ d_U~~ La~ prueba~ no~ concluye$

### **Test de Kolmogorov-Smirnov (Prueba de Normalidad)**

**Ho:** Los residuos no siguen una distribución normal

**Ha:** Los residuos siguen una distribución normal
