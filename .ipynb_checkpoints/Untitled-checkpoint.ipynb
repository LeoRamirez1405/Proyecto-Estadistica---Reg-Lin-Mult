```python
#comprobando los supuestos

import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("2022 MLB Player Stats - Pitching.csv",sep = ';')

X = data[["IP","BF"]] # variables independientes
y = data["BB"] # variable dependiente

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()


plt.scatter(model.predict(), model.resid)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
# -----errores independientes-----------
# Calculate the Durbin-Watson statistic
dw = sm.stats.stattools.durbin_watson(model.resid)

# Print the Durbin-Watson statistic
print('Durbin-Watson statistic:', dw)
#Si el valor estadístico de Durbin-Watson está cerca de 2,
#entonces no hay una autocorrelación significativa en los residuos
#y los errores pueden considerarse independientes.


```

```python
#-------valor esperado del error aleatorio e_i es 0----------------
print(model.summary())
# el coeficiente de la constante es 1.2546 y su valor p es menor que 0.05,
# lo que indica que es significativamente diferente de cero. 
# Por lo tanto, no se cumple el supuesto de que
# el valor esperado del error aleatorio es cero para este modelo de regresión.
```

```python
#----------Homocedasticidad----------------------
import statsmodels.stats.diagnostic as smd
# Realizar la prueba de Breusch-Pagan
bp_test = smd.het_breuschpagan(model.resid, X)

# Imprimir los resultados de la prueba de Breusch-Pagan
print('Breusch-Pagan test:')
print('LM statistic: {:.4f}'.format(bp_test[0]))
print('p-value: {:.4f}'.format(bp_test[1]))
print('F statistic: {:.4f}'.format(bp_test[2]))
print('F p-value: {:.4f}'.format(bp_test[3]))

# Realizar la prueba de White
white_test = smd.het_white(model.resid, X)

# Imprimir los resultados de la prueba de White
print('\nWhite test:')
print('LM statistic: {:.4f}'.format(white_test[0]))
print('p-value: {:.4f}'.format(white_test[1]))
print('F statistic: {:.4f}'.format(white_test[2]))
print('F p-value: {:.4f}'.format(white_test[3]))

# Si el valor p de cualquiera de las pruebas es menor que 0.05,
# entonces se puede concluir que no se cumple el supuesto de homocedasticidad. 
# Si el valor p es mayor que 0.05, entonces se puede concluir que
# se cumple el supuesto de homocedasticidad.
```

```python
#---------------errores siguen distribucion normal con media 0 y varianza constante-------------
# Realizar la prueba de Shapiro-Wilk
shapiro_test = smd.normal_ad(model.resid)

# Imprimir los resultados de la prueba de Shapiro-Wilk
print('Shapiro-Wilk test:')
print('statistic: {:.4f}'.format(shapiro_test[0]))
print('p-value: {:.4f}'.format(shapiro_test[1]))

# Si el valor p es mayor que 0.05, 
# entonces se puede concluir que los errores se distribuyen normalmente. 
print("")
# Imprimir los resultados de la prueba de Breusch-Pagan
print('Breusch-Pagan test:')
print('LM statistic: {:.4f}'.format(bp_test[0]))
print('p-value: {:.4f}'.format(bp_test[1]))
print('F statistic: {:.4f}'.format(bp_test[2]))
print('F p-value: {:.4f}'.format(bp_test[3]))

# Si el valor p es mayor que 0.05, entonces se puede concluir
# que la varianza de los errores es constante. 

# Si ambas pruebas se cumplen, entonces se puede concluir que
# se cumple el supuesto de que los errores siguen una distribución
# normal con media 0 y varianza constante en la regresión lineal múltiple.
```

```python
# representacion grafica de los residuos, su distribucion, esperanza y varianza

sns.histplot(model.resid, kde=True)
plt.show()
sm.qqplot(model.resid, line='s')
plt.show()

_, pval, _, _ = het_breuschpagan(model.resid, X)
print("P-Value:",pval,"\n")

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
print(vif)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as sc
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson


data = pd.read_csv("2022 MLB Player Stats - Pitching.csv", sep=';')
data = data[data['Lg'] == 'AL']
data = data[data['Tm'].isin(['OAK', 'SEA', 'TEX','NYY'])]
data = data.rename(columns={'SO/W': 'SO_W'})

oak = data['SO_W'][data['Tm'] == 'OAK']
sea = data['SO_W'][data['Tm'] == 'SEA']
tex = data['SO_W'][data['Tm'] == 'TEX']
nyy = data['SO_W'][data['Tm'] == 'NYY']

var = [oak,sea,tex,nyy]

model = ols('SO_W ~ Tm', data=data).fit()

#print(var)
print(sc.f_oneway(*var))

#    supuestos
# 1ro - errores distribucion normal con media 0
print(sc.shapiro(data['SO_W']))
# 2do - errores independientes
residuos = model.resid
print(durbin_watson(model.resid))

# si d es aprox 2 ent son independientes
# si d es mayor q 4 o menor q 0 no se puede decir

# 3ro - residuos igual varianza
sc.levene(*var)

from statsmodels.graphics.gofplots import qqplot
qqplot(residuos, line='s')
plt.show()
sns.histplot(residuos, kde=True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("2022 MLB Player Stats - Pitching.csv", sep=';')
data = data[data['Lg'] == 'AL']
data = data.drop(['Rk', 'Name','Age','Tm','Lg'], axis=1)


# Estandarizamos los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

#Hacemos la reduccion de dimensiones
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
plt.show()

#Aplicamos el algoritmo K-Means para agrupar los datos en 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit_predict(data_pca)

# Graficamos los clusters
plt.scatter(data_pca[:,0], data_pca[:,1], c=clusters, cmap='viridis')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```
